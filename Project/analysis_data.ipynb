```python
import pandas as pd
```

```python
sp500_filtered = pd.read_csv("sp500_filtered.csv", parse_dates=["date"])
print(sp500_filtered.shape)
sp500_filtered.head()
```

```python
# =========================
# FF5
# =========================
ff5 = pd.read_csv(
    "factor data/F-F_Research_Data_5_Factors_2x3_daily.csv",
    skiprows=3
)
ff5 = ff5.rename(columns={ff5.columns[0]: "date"})
ff5 = ff5[ff5["date"].astype(str).str.isnumeric()]
ff5["date"] = pd.to_datetime(ff5["date"], format="%Y%m%d")

# =========================
# Momentum
# =========================
mom = pd.read_csv(
    "factor data/F-F_Momentum_Factor_daily.csv",
    skiprows=13
)
mom = mom.rename(columns={mom.columns[0]: "date"})
mom = mom[mom["date"].astype(str).str.isnumeric()]
mom["date"] = pd.to_datetime(mom["date"], format="%Y%m%d")

# =========================
# Industry 49
# =========================
ind49 = pd.read_csv(
    "factor data/49_Industry_Portfolios_Daily.csv",
    skiprows=9
)
ind49 = ind49.rename(columns={ind49.columns[0]: "date"})
ind49 = ind49[ind49["date"].astype(str).str.isnumeric()]
ind49["date"] = pd.to_datetime(ind49["date"], format="%Y%m%d")
```

```python
#Trim everything starting from 2005.
start_date = "2005-01-01"
end_date = "2026-02-27"

sp500_filtered = sp500_filtered[
    (sp500_filtered["date"] >= start_date) & 
    (sp500_filtered["date"] <= end_date)
]

ff5 = ff5[
    (ff5["date"] >= start_date) & 
    (ff5["date"] <= end_date)
]

mom = mom[
    (mom["date"] >= start_date) & 
    (mom["date"] <= end_date)
]

ind49 = ind49[
    (ind49["date"] >= start_date) & 
    (ind49["date"] <= end_date)
]
```

```python
# Use the most recent 20 years' data
print("SP500:", sp500_filtered ["date"].min(), sp500_filtered ["date"].max())
print("FF5:", ff5["date"].min(), ff5["date"].max())
print("MOM:", mom["date"].min(), mom["date"].max())
print("IND49:", ind49["date"].min(), ind49["date"].max())
```

```python
sp500 = sp500_filtered.copy()

sp500["ret"] = sp500.groupby("ticker")["adj_close"].pct_change()

sp500["y"] = sp500.groupby("ticker")["ret"].shift(-1)
```

```python
ind49
```

```python
sp500 = sp500.merge(ff5, on="date", how="left")
sp500 = sp500.merge(mom, on="date", how="left")
```

```python
sp500["mom_5"] = sp500.groupby("ticker")["ret"].rolling(5).mean().reset_index(0, drop=True)
sp500["vol_20"] = sp500.groupby("ticker")["ret"].rolling(20).std().reset_index(0, drop=True)
```

```python
sp500.columns
```

```python
feature_cols = [
    "mom_5", "vol_20",
    "Mkt-RF", "SMB", "HML", "RMW", "CMA", "Mom"
]
```

```python
#Constructing the Covariance Matrix
returns_pivot = sp500.pivot(index="date", columns="ticker", values="ret")

V_dict = {}
window = 60

for i in range(window, len(returns_pivot)):
    date = returns_pivot.index[i]
    V_dict[date] = returns_pivot.iloc[i-window:i].cov()
```

```python
#Constructing IPO Data
data = []

for date in sp500["date"].unique():
    df_t = sp500[sp500["date"] == date]

    if date not in V_dict:
        continue

    df_t = df_t.dropna(subset=["y"] + feature_cols)

    if len(df_t) < 50:
        continue

    tickers = df_t["ticker"]

    X = df_t[feature_cols].values
    y = df_t["y"].values
    V = V_dict[date].loc[tickers, tickers].values

    data.append((X, y, V))
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

def solve_mvo(mu, V, lam=1.0):
    return np.linalg.solve(lam * V, mu)

results = []

for X, y, V in data[:200]:  # 先跑一点
    model = LinearRegression().fit(X, y)
    mu = model.predict(X)

    z = solve_mvo(mu, V)
    ret = z @ y

    results.append(ret)

print("Mean return:", np.mean(results))
```
